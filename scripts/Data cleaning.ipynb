## Get country data from lat/long
https://www.geeksforgeeks.org/get-the-city-state-and-country-names-from-latitude-and-longitude-using-python/

```python
import certifi
import ssl
import geopy.geocoders
from geopy.geocoders import Nominatim
ctx = ssl.create_default_context(cafile=certifi.where())
geopy.geocoders.options.default_ssl_context = ctx

geolocator = Nominatim(scheme='http', user_agent="geoapiExercises")
```

```python
biden_data_country = new_biden_data
```

```python
from geopy.exc import GeocoderTimedOut
```

```python
searched_location = {}
```

```python
for item, frame in biden_data_country['country'].iteritems():
    if pd.isnull(frame):
        lat = biden_data_country['lat'][item]
        long = biden_data_country['long'][item]
        lat_and_long = str(lat) + ", " + str(long)
        if lat_and_long in searched_location:
            biden_data_country['country'][item] = searched_location[lat_and_long].raw['address']['country']
        else:
            try:
                location = geolocator.reverse(lat_and_long, language="en", timeout=None)
                if location != None:
                    biden_data_country['country'][item] = location.raw['address']['country']
                    searched_location[lat_and_long] = location
            except GeocoderTimedOut as e:
                continue
```

```python
biden_data_country.to_csv('clean_biden2.csv', index=False)  
```

# Combine Biden and Trump Data
Add a new column 'data_origin' whose value is only 'Trump', 'Biden', or 'Both'

```python
# Find intersection between Biden's and Trump's data
intersect_data = pd.merge(biden_data_hashtag, trump_data_hashtag, how = 'inner', on = ['created_at', 'tweet', 'user_screen_name'])
```

```python
intersect_data.head()
```

```python
# Delete duplicate columns
del intersect_data['likes_y']
del intersect_data['retweet_count_y']
del intersect_data['source_y']
del intersect_data['user_name_y']
del intersect_data['user_description_y']
del intersect_data['user_followers_count_y']
del intersect_data['user_location_y']
del intersect_data['lat_y']
del intersect_data['long_y']
del intersect_data['city_y']
del intersect_data['country_y']
del intersect_data['continent_y']
del intersect_data['state_y']
del intersect_data['state_code_y']
del intersect_data['hashtags_y']
```

```python
# rename the columns to remove the '_x'
intersect_data.columns = ['created_at', 
                          'tweet',
                          'likes',
                          'retweet_count',
                          'source',
                          'user_name',
                          'user_screen_name',
                          'user_description',
                          'user_followers_count',
                          'user_location',
                          'lat',
                          'long',
                          'city',
                          'country',
                          'continent',
                          'state',
                          'state_code',
                          'hashtags']
```

```python
# make a set of the intersecting data consisting of (created_at, tweet, user_screen_name)
intersect_set = set()
for item, frame in intersect_data['tweet'].iteritems():
    intersect_set.add((intersect_data['created_at'][item], intersect_data['tweet'][item], intersect_data['user_screen_name'][item]))
```

```python
# Traverse Biden's data to add the new column
biden_data_origin = []
total_intersect = 0 # for cross-checking
for item, frame in biden_data_hashtag['tweet'].iteritems():
    cur_tuples = (biden_data_hashtag['created_at'][item], biden_data_hashtag['tweet'][item], biden_data_hashtag['user_screen_name'][item])
    if cur_tuples in intersect_set:
        total_intersect += 1
        biden_data_origin.append('Both')
    else:
        biden_data_origin.append('Biden')
```

```python
biden_data_hashtag['data_origin'] = biden_data_origin
```

```python
biden_data_hashtag.head()
```

```python
combined_data = biden_data_hashtag
```

```python
trump_hashtag_copy = trump_data_hashtag.copy(deep=True)
```

```python
# Traverse Trump's data
trump_data_origin = []
drop_rows = []
for item, frame in trump_hashtag_copy['tweet'].iteritems():
    cur_tuples = (trump_hashtag_copy['created_at'][item], trump_hashtag_copy['tweet'][item], trump_hashtag_copy['user_screen_name'][item])
    if cur_tuples in intersect_set:
        # drop the rows
        drop_rows.append(item)
    else:
        trump_data_origin.append('Trump')
        
```

```python
trump_hashtag_copy = trump_hashtag_copy.drop(drop_rows)
```
